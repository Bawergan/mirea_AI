```python
#4.3.1

import matplotlib.pyplot as plt
import random
import numpy as np

size = 100
data = [random.uniform(0, 1) for i in range(size)]
mean = np.mean(data)
median = np.median(data)
print('mean -', mean)
print('median -', median)
plt.scatter(np.arange(0,size), data)
plt.plot([0, size], [mean, mean], label = 'mean')
plt.plot([0, size], [median, median], label='median')
plt.legend()
```

```python
#4.3.2

import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0.0, 17, 0.1)
y = np.sqrt(1 + np.power(np.exp(1),np.sqrt(x)) + np.cos(x**2))/(np.abs(1 - np.power(np.sin(x),3))) + np.log(np.abs(2 * x))

plt.scatter(x[:len(x)//2], y[:len(x)//2])
plt.plot(x[len(x)//2:], y[len(x)//2:],c = 'r')
plt.grid()
```
